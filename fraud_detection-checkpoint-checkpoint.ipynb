```python
#Selecting relevant data from raw data by filtering on specific transaction types
df1 = df[(df['type'] == 'TRANSFER') | (df['type'] == 'CASH_OUT')].copy()
#Removing unnecessary columns from the selected data
df1.drop(['step', 'nameOrig', 'nameDest', 'isFlaggedFraud'], axis=1, inplace=True)
#Resetting the index of the data
df1 = df1.reset_index(drop=True)

#Encoding the 'type' column as a numerical category using LabelEncoder
type_label_encoder = preprocessing.LabelEncoder()
type_category = type_label_encoder.fit_transform(df1['type'].values)
df1['typeCategory'] = type_category
#Printing the first five rows of the processed data
print(df1.head())
```

```python
#Defining the feature names to be used in the model
feature_names = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'typeCategory']
#Selecting the features and the target variable from the selected data
X = df1[feature_names]
y = df1['isFraud']
#let's look at the data 
print(X.head())
print(y.head())
```

```python
#Count the number of fraudulent records in the dataset
number_records_fraud = len(df1[df1['isFraud'] == 1])
#Get the index of all fraudulent records

fraud_indices = df1[df1['isFraud'] == 1].index.values
#Get the index of all non-fraudulent records
nonfraud_indices = df1[df1['isFraud'] == 0].index
#Randomly select non-fraudulent records to match the number of fraudulent records
random_nonfraud_indices = np.random.choice(nonfraud_indices, number_records_fraud, replace=False)
random_nonfraud_indices = np.array(random_nonfraud_indices)

#Combine the indices of fraudulent and non-fraudulent records for undersampling
under_sample_indices = np.concatenate([fraud_indices, random_nonfraud_indices])
#Get the undersampled data by selecting only the records with the combined indices
under_sample_data = df1.iloc[under_sample_indices, :]
#Separate the features and the target variable from the undersampled data                          
                          
X_undersample = under_sample_data[feature_names].values
y_undersample = under_sample_data['isFraud'].values


print("non-fraudulent records: ", len(under_sample_data[under_sample_data['isFraud'] == 0]) / len(under_sample_data))
print("fraudulent records: ", len(under_sample_data[under_sample_data['isFraud'] == 1]) / len(under_sample_data))
print("Number of undersampled records: ", len(under_sample_data))
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

#Split the undersampled dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_undersample, y_undersample, test_size=0.3, random_state=0)
#Train a logistic regression model on the training set
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
#Get the predicted scores for the testing set
y_pred_score = lr_model.decision_function(X_test)
#Compute the ROC curve and AUC score for the logistic regression model
fpr, tpr, thresholds = roc_curve(y_test, y_pred_score)
roc_auc = auc(fpr,tpr)

# Plot the ROC curve with AUC score and the random guessing line
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.0])
plt.ylim([-0.1,1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
